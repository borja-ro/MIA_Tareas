# Introducción a las Series Temporales

En este cuaderno, exploraremos los conceptos básicos de las series temporales, su importancia en el análisis de datos y cómo visualizarlas.

## 1. ¿Qué es una Serie Temporal?

Una **serie temporal** es una secuencia de observaciones recogidas en intervalos regulares a lo largo del tiempo. Estas observaciones pueden ser cualquier tipo de dato que varíe con el tiempo, como:
- Datos climáticos (temperatura, precipitación).
- Datos económicos (ventas mensuales, precios de acciones).
- Datos de tráfico web.
- Datos de sensores en dispositivos IoT.

### Características de las Series Temporales
1. **Tendencia:** Cambio gradual a largo plazo (ej. aumento de temperatura global).
2. **Estacionalidad:** Patrones que se repiten en intervalos fijos (ej. ventas en Navidad).
3. **Ciclos:** Fluctuaciones que no tienen una periodicidad fija (ej. ciclos económicos).
4. **Ruido:** Variaciones aleatorias que no siguen un patrón claro.

### Ejemplo de Series Temporales
Vamos a ver algunos ejemplos de series temporales en diferentes contextos:
- **Clima:** Temperaturas diarias en una ciudad.
- **Economía:** Precios mensuales de una acción en bolsa.
- **Salud:** Número de casos diarios de una enfermedad.
- **Tecnología:** Tráfico de datos en una red por hora.

## 2. Importancia de las Series Temporales

Las series temporales son fundamentales en muchos campos, como:
- **Predicción:** Predecir valores futuros basados en datos históricos (ej. demanda de productos).
- **Monitoreo:** Detectar anomalías en tiempo real (ej. fraudes en transacciones bancarias).
- **Planificación:** Optimizar recursos basados en patrones históricos (ej. inventario en tiendas).
- **Investigación:** Analizar tendencias y patrones en datos científicos (ej. cambio climático).

## 3. Visualización de Series Temporales

La visualización es una herramienta clave para entender las series temporales. Vamos a cargar y visualizar algunos ejemplos utilizando Python y las bibliotecas `pandas` y `matplotlib`.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Ejemplo 1: Temperatura mensual
data_temp = {
    'Mes': ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic'],
    'Temperatura': [12, 14, 16, 18, 22, 25, 28, 29, 26, 22, 18, 14]
}

df_temp = pd.DataFrame(data_temp)

plt.figure(figsize=(10, 6))
plt.plot(df_temp['Mes'], df_temp['Temperatura'], marker='o', label='Temperatura')
plt.title('Temperatura Mensual')
plt.xlabel('Mes')
plt.ylabel('Temperatura (°C)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Ejemplo 2: Ventas mensuales de un producto
data_ventas = {
    'Mes': ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic'],
    'Ventas': [120, 150, 130, 160, 200, 180, 220, 250, 210, 190, 170, 300]
}

df_ventas = pd.DataFrame(data_ventas)

plt.figure(figsize=(10, 6))
plt.plot(df_ventas['Mes'], df_ventas['Ventas'], marker='o', color='green', label='Ventas')
plt.title('Ventas Mensuales')
plt.xlabel('Mes')
plt.ylabel('Ventas (unidades)')
plt.grid(True)
plt.legend()
plt.show()
```

## 4. Componentes de una Serie Temporal

Las series temporales suelen descomponerse en varios componentes:
1. **Tendencia (Trend):** Cambio gradual a largo plazo.
2. **Estacionalidad (Seasonality):** Patrones que se repiten en intervalos fijos.
3. **Ciclos (Cycles):** Fluctuaciones que no tienen una periodicidad fija.
4. **Ruido (Noise):** Variaciones aleatorias.

### Ejemplo de Descomposición
Vamos a descomponer una serie temporal en sus componentes utilizando `statsmodels`.

```python
from statsmodels.tsa.seasonal import seasonal_decompose
import numpy as np

# Crear una serie temporal de ejemplo con tendencia y estacionalidad
np.random.seed(0)
time = np.arange(0, 200, 1)
trend = 0.1 * time
seasonality = 10 * np.sin(time / 5)
noise = np.random.normal(0, 2, len(time))
series = trend + seasonality + noise

# Descomponer la serie
decomposition = seasonal_decompose(series, period=10)

# Graficar los componentes
plt.figure(figsize=(12, 8))
decomposition.plot()
plt.show()
```

## Análisis de la acción de Microsoft (MSFT)

**Dataset elegido:** Precio diario de la acción de Microsoft (MSFT) entre abril de 2015 y marzo de 2021, obtenido del archivo `Microsoft_Stock.csv`.

Esto es un ejemplo típico de TimeSeries en actividad financiera. El dataset es este: https://www.kaggle.com/datasets/vijayvvenkitesh/microsoft-stock-time-series-analysis

### Paso 1 – Carga y visualización de los datos

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Cargar los datos
df = pd.read_csv('Microsoft_Stock.csv', parse_dates=['Date'], dayfirst=False)
df = df.sort_values('Date')
df.set_index('Date', inplace=True)

print(f"fechas desde hasta: {df.index.min().date()} — {df.index.max().date()}")
print(f"datos observados: {len(df)}")
print(df.head())

# Visualizar el precio de cierre
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Close'], color='steelblue', linewidth=0.8)
plt.title('Precio de Cierre Diario de Microsoft (MSFT)')
plt.xlabel('Fecha')
plt.ylabel('Precio de Cierre (USD)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Paso 2 – Análisis de la serie temporal

La serie muestra una tendencia claramente al alza a lo largo de los seis años analizados, con el precio de cierre pasando de unos 40 USD en 2015 a cerca de 236 USD en 2021, lo que supone un crecimiento aproximado de x6, interrumpido únicamente por la caida de 2020 asociada al impacto inicial de la pandemia de COVID-19. después la recuperación es rápida y constante, No se aprecia una estacionalidad fuerte en frecuencia diaria, aunque al agregar los datos mensualmente pueden intuirse patrones ( > actividad en cierres de trimestre). En total, la serie no es estacionaria, ya que su media crece de forma sostenida a lo largo del tiempo.

### Paso 3 – Descomposición de la serie

Resampleamos a frecuencia mensual (media del precio de cierre) para aplicar `seasonal_decompose` con periodo = 12 meses.

```python
# Resamplear a frecuencia mensual (media del precio de cierre)
df_monthly = df['Close'].resample('ME').mean()

# Descomposición aditiva con periodo de 12 meses
decomposition = seasonal_decompose(df_monthly, model='additive', period=12)

fig = decomposition.plot()
fig.set_size_inches(12, 8)
fig.suptitle('Descomposición de la Serie Temporal – Microsoft (MSFT)', y=1.02, fontsize=14)
plt.tight_layout()
plt.show()
```

### Paso 4 – Informe de resultados

#### Patrones observados
- **Tendencia creciente sostenida:** El precio de Microsoft pasó de ~40 USD en abril de 2015 a ~236 USD en marzo de 2021. La tendencia extraída por la descomposición confirma este crecimiento prácticamente continuo.
- **Impacto del COVID-19:** En marzo de 2020 se produce una caída brusca (~30 %) seguida de una recuperación muy rápida, visible tanto en la serie original como en el componente residual.
- **Estacionalidad débil:** El componente estacional muestra oscilaciones de poca amplitud (pocos dólares) comparadas con la magnitud de la tendencia. Esto es esperable en precios de acciones, donde no existe un patrón estacional rígido como en datos climáticos o de ventas minoristas.
- **Ruido considerable:** El residuo presenta picos notables, especialmente alrededor de 2020, lo que indica que la volatilidad del mercado introduce variaciones difíciles de capturar con un modelo puramente aditivo.

#### Eventos externos relevantes
- **2018:** Crecimiento acelerado por la expansión de Azure y los servicios en la nube.
- **Marzo 2020:** Desplome por la pandemia de COVID-19.
- **2020-2021:** Recuperación impulsada por el auge del teletrabajo y la digitalización.

#### Estacionariedad
La serie original **no es estacionaria** (la media y la varianza cambian con el tiempo). Para modelar esta serie con técnicas como ARIMA sería necesario aplicar al menos una diferenciación.

#### Posibles mejoras
- Aplicar un modelo **ARIMA/SARIMA** sobre la serie diferenciada para realizar predicciones.
- Utilizar una descomposición **multiplicativa** en lugar de aditiva, ya que la amplitud de las fluctuaciones crece con el nivel de precios.
- Incorporar variables exógenas (índices bursátiles, tipos de interés) mediante modelos **SARIMAX**.
